OOP QUIZ Q4: WHAT IS INHERITANCE? 
1. The property of OOP that allows child classes to be created with parameters and methods from parent (super) classes
2. A property of OOP that results from getting data from one data object and passing it to another
3. A property of OOP that results in functions descending from other functions

![SegmentLocal](Types-of-OOPS-2.gif "segment")

```python
#1. The property of OOP that allows child classes to be created with parameters and methods from parent (super) classes
#inheritance
#example:

#here we define a calss of muppet with attribute of name
class Muppet:
    def __init__(self, name):
        self.name = name
    def display_name(self):
        print(f"Muppet's name: {self.name}")

#here we define a child class of muppet, frog, with attribute of mantra
class Frog(Muppet):
    def mantra(self):
        print("It's not easy being green")

#the child class maintains its own attribute, plus the attribute of its parent class
x = Frog("Kermit")
x.display_name()
x.mantra()
```

```python
#to fix it, let's define texture in the class after "fill_state"
class crate(object):
   def __init__(self, weight, dimensions, fill_state, texture):
      self.weight = weight
      self.dimensions = dimensions
      self.fill_state = fill_state
      self.texture = texture
   def fill(self):
      if self.fill_state==0:
         self.fill_state=1
      else:
         return ('No room!')
c = crate(weight = 40, dimensions = (15,10,10), fill_state = 0, texture = 'wooden')

#so now if we run fill, it will fill the crate
c.fill() #fills it
print(c.fill_state) #we can see that it is full
```

```python
print(c.fill()) #no more space in our crate
```
